### Description of Feature Extraction

To capture mutation-driven resistance more directly and improve the temporal generalization of our models, we can leverage two primary methods for extracting features from raw genome sequences: **SNP Feature Extraction** and **K-mer Feature Extraction**.

#### 1. SNP (Single Nucleotide Polymorphism) Feature Extraction

This process involves identifying specific, known resistance-conferring mutations at the single nucleotide level within target genes. For instance, well-documented mutations in the `gyrA` gene are known to confer resistance to fluoroquinolone antibiotics, while mutations in `rpoB` are associated with rifampicin resistance. The extraction process typically involves:

*   **Alignment**: Aligning the raw genome sequences of isolates to a reference genome or specific resistance genes.
*   **Variant Calling**: Identifying single nucleotide variations (SNVs) or small insertions/deletions (indels) compared to the reference or wild-type sequence.
*   **Feature Representation**: These identified mutations can then be represented as binary features: `1` if a specific resistance-conferring mutation is present, and `0` if it is absent. In cases where multiple distinct mutations or alleles exist within a gene, these could also be encoded as categorical features.

#### 2. K-mer Feature Extraction

K-mer extraction involves breaking down raw genome sequences into all possible overlapping short sequences of a fixed length `k`. For example, a sequence `AGCTAGC` with `k=3` would yield k-mers `AGC`, `GCT`, `CTA`, `TAG`, `AGC`. The process entails:

*   **K-mer Generation**: Iterating through each genome sequence and extracting all k-mers of the specified length `k`.
*   **Feature Representation**: These k-mers can then be used to form numerical features. Common approaches include:
    *   **Presence/Absence**: Creating binary features where `1` indicates the presence of a specific k-mer in an isolate's genome, and `0` indicates its absence.
    *   **Counts**: Recording the frequency of each unique k-mer within a genome.
    *   **Vectorization**: More advanced techniques like TF-IDF (Term Frequency-Inverse Document Frequency) can be applied to k-mer counts to weigh their importance, similar to natural language processing.

Both **SNP features** and **K-mer features** are derived directly from the fundamental genomic sequences. They are crucial because they allow us to capture fine-grained, mutation-driven resistance mechanisms that might not be evident from gene presence/absence alone. By including these features, we can build models that are more robust and better able to generalize to newly emerging resistant strains over time, thereby improving temporal prediction accuracy.

#### Strategy: Broadening Features to the Pangenome

1.  **What is a Pangenome?**
    The pangenome of a bacterial species refers to the **total set of genes** found across all strains within that species. It comprises a 'core genome' (genes present in all strains) and an 'accessory genome' (genes present in some, but not all, strains).

2.  **Why Consider This Approach?**
    This advanced strategy would be pursued if initial, more targeted feature sets (such as SNP-based features or K-mer signatures), despite their sophistication, prove insufficient for achieving robust predictive performance and, critically, for demonstrating strong temporal generalizability. 3.  **Methodology: Gene Presence/Absence Matrix for All Genes**
    Instead of focusing only on known resistance genes or specific SNP loci, this strategy involves constructing a **gene presence/absence matrix for *all* genes** within the species' pangenome across all isolates. Each column in this matrix would represent a unique gene in the pangenome, and each row an isolate, with values indicating whether that gene is present (1) or absent (0) in the isolate.

4.  **Goal: Discovering Novel or Non-Traditional Markers**
    By including the entire pangenome as features, the machine learning model can potentially discover **novel or non-traditional genetic markers** that correlate with resistance phenotypes. These might include genes not currently recognized as direct AMR determinants (e.g., genes involved in metabolism, stress response, or virulence that indirectly influence resistance) or combinations of accessory genes that form a predictive signature for resistance.
